```python
Main_file_path = Path.cwd().parent
Data_file_path = Main_file_path / 'data'
#Data_file = Data_file_path / 'Basic_Macro_Indicators.csv'
Data_file = Data_file_path / 'MacroSEI.csv'
```

```python
#DIR_DATA =  Path.cwd().parents[1].joinpath("DSCBI_PROJECT", "data")
#FILE = DIR_DATA.joinpath("Basic_Macro_Indicators.csv")
# ==============================
def load_data(Data_file):
    data = pd.read_csv(Data_file)
    return data
# ==============================
data = load_data(Data_file)
```

```python
data.head()
```

```python
data = data[0:66]
```

```python
#drop all unnamed columns
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
#melting the dataframe
data_melted = data.melt(id_vars=['Sector', 'Indicator'], var_name='Year', value_name='Value')
```

```python
# search the Value column in df_melted if whether the value contains a comma and remove it and also a percent sign if any
data_melted['Value'] = data_melted['Value'].apply(lambda x: re.sub(r'[,%]', '', str(x)) if isinstance(x, str) else x)

data_melted['Year'] = data_melted['Year'].apply(lambda x: re.sub(r'[ Proj]', '', str(x)) if isinstance(x, str) else x)

data_melted['Value'] = pd.to_numeric(data_melted['Value'], errors='coerce')
# ==============================
# ==============================
```
